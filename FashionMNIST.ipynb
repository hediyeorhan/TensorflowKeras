```
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```
mnist = tf.keras.datasets.fashion_mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train / 255.0 # 1 ve 0 arasına sıkıştırmak için
X_test = X_test / 255.0

class mycallbacks(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs = {}):
    if(logs.get("accuracy") > 90):
      self.model.stop_training

callbacks = mycallbacks()

model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28, 28)),
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])

model.compile(loss="sparse_categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
model.fit(X_train, y_train, epochs = 8, callbacks = [callbacks])

loss = model.evaluate(X_test, y_test)
```

```
mnist = tf.keras.datasets.fashion_mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(60000, 28, 28, 1)
X_train = X_train / 255.0

X_test = X_test.reshape(10000, 28, 28, 1)
X_test = X_test / 255.0

model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(32, (3,3), activation="relu", input_shape=(28, 28, 1)),
                                    tf.keras.layers.MaxPooling2D(2, 2),
                                    tf.keras.layers.Conv2D(64, (3,3), activation="relu"),
                                    tf.keras.layers.MaxPooling2D(2, 2),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(128, activation="relu"),
                                    tf.keras.layers.Dense(10, activation="softmax")])

model.compile(optimizer="adam", loss = "sparse_categorical_crossentropy", metrics=["accuracy"])
model.fit(X_train, y_train, epochs = 8)

model.summary()

loss = model.evaluate(X_test, y_test)
```
